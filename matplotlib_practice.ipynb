```python
#import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create plot
plt.plot(x_axis, y_axis)
```

```python
#create plot with object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create plot and add label for legend
plt.plot(x_axis, y_axis, marker="d", color="blue", linewidth=2, label='Boston')
#create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
#set y limit between 0 and 45
plt.ylim(0,45)
#create title
plt.title("PyBer Fare by Month")
#add grid
plt.grid()
#add legend
plt.legend()
```

```python
#create plot
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#create plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
#invert y axis
plt.gca().invert_yaxis()
# Add the legend.
plt.legend()
```

```python
#create plot w/ ax.plt
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label= 'Chicago')
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
#
plt.scatter(y_axis, x_axis,s= [i*3 for i in y_axis], c="red", label='Chicago')
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
fig,ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], c="skyblue", alpha=.8, edgecolor="k", linewidth=2, label='Boston')
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    shadow=True,
    counterclock=False,
    startangle=90,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
#get standard deviation
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator
#increase plot fig size
fig, ax = plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))

#create minor ticks at increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
